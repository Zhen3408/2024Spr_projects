# Expansion on Hypothesis 3 (Continue)
## The Unemployment Rate detail about countries with relatively high employment-education equity in the European Countries
Using standard deviation as a measure to show how spread out the unemployment rates are for different education levels within each country. A smaller standard deviation would indicate that the unemployment rates are more uniform across different education levels, suggesting higher equity.

```python
# Group by 'Country' and calculate the standard deviation of the 'Unemployment Rate' for each 'Country'
std_devs = df_oecd_education.groupby('Country')['Unemployment Rate'].std().reset_index()
# Sort the results by standard deviation in ascending order to get the countries with the most equity
sorted_countries = std_devs.sort_values(by='Unemployment Rate')
# Select the top four countries
top_countries = sorted_countries.head(4)['Country'].tolist()
print(f"The top four Europe countries with the most equity in unemployment rates by education level are: {top_countries}")

df_oecd = pd.read_csv('./Data/OECD/EducationalAttainment.csv')
df_oecd_top_countries = process_oecd_specific_countries(df_oecd, top_countries) 

for country in top_countries:
    df_oecd_top_countries.loc[country].plot(figsize=(15, 7), linewidth = 1.5, marker='o')
    plt.grid()
    plt.xlabel('Date')
    plt.ylabel('Unemployment Rate (%)')
    plt.title(f'Unemployment Rate in {country} by Education Level')
    plt.ylim(0, 15)
    plt.show()
```

## Observation
The plot shows that the unemployment rate is relatively low for all education levels in those countries. Even though among those countries, the unemployment rate is still related to the education level. As we can see most of the time, the unemployment rate is always higher for those with less than a high school diploma. One exception is Iceland, there shows no obvious unemployment rate difference among different education level groups.

Also, the unemployment rate in those countries did not show a significant fluctuation during the COVID-19 pandemic, which is different from the United States. Seems like the job market circumstance in those countries is more stabled during the pandemic.
